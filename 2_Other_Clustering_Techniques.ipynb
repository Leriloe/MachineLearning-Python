## Segmentación de Precio de Acciones
<p>En este capítulo vamos a entender y visualizar una segmentacion jerarquica. Para eso vamos a analizar los datos de los precios de acciones de algunas compañias importantes</p>

```python
# Importa Pandas y Matplotlib
import pandas as pd
import matplotlib.pyplot as plt


# Importa el archivo "Company Stock Price Movements.csv"
df = pd.read_csv('Company Stock Price Movements.csv')
```

```python
# Análisis Exploratorio

df.head()
```

## Hierarchical Clustering
<p>Antes de crear los segmentos, vamos a procesar los datos para sacar el máximo valor del algoritmo</p>
<p><img src="https://www.statisticshowto.com/wp-content/uploads/2016/11/clustergram.png"></p>

```python
# Crea una variable con el nombre de las empresas
companies = df['Company'].values
```

```python
# Crea una variable con los valores de las acciones
movements = df.drop('Company', axis=1).values
```

```python
# Importa modulos necesarios
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import normalize

# Normaliza los movimientos
norm_movements = normalize(movements)

# Calcula la distancia por el metodo 'complete'
mergings = linkage(norm_movements,method='complete')

# Grafica el dendograma
f = plt.figure(figsize=(18,8))
dendrogram(mergings,
           labels=companies,
           leaf_rotation=90,
           leaf_font_size=14,)
plt.show()
```

## Gaussian Mixture Model
<p>Existen muchos más algoritmos de segmentación, de los cuales vamos a mencionar el <b>Gaussian Mixture Model (GMM)</b>. El cual es un modelo probabilístico en el que se considera que las observaciones siguen una distribución probabilística formada por la combinación de múltiples distribuciones normales (componentes). Puede entenderse como una generalización de K-means con la que, en lugar de asignar cada observación a un único cluster, se obtiene una distribución probabilidad de pertenencia a cada uno. 

Para conocer más detalles puedes consultar la documentación en el siguiente <a href="https://scikit-learn.org/stable/modules/mixture.html#:~:text=A%20Gaussian%20mixture%20model%20is,Gaussian%20distributions%20with%20unknown%20parameters.">Link</a></p>

```python
# Importa los datos de los archivos mv1, mv2, mv3 y mv4
mv1 = pd.read_csv('mv1.csv', index_col = 0)
mv2 = pd.read_csv('mv2.csv', index_col = 0)
mv3 = pd.read_csv('mv3.csv', index_col = 0)
mv4 = pd.read_csv('mv4.csv', index_col = 0)
```

```python
# Vamos a visualizar en un grafico de dispersión los segmentos de cada conjunto de datos
fig, (axs) = plt.subplots(2,2, figsize = (6,6))

for ax, df in zip(axs.flatten(), [mv1, mv2, mv3, mv4]):
    for cat in df['cat'].unique():
        ax.scatter(df[df.cat==cat].x, df[df.cat==cat].y, label=cat, alpha=.15)
    ax.legend()
```
